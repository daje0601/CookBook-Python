```python
# 혼용되는 용어 통일을 위한 object 추출 
df1.select_dtypes(exclude="number").columns
```

```python
# 혼용되는 용어 통일을 위한 object 리스트화
object_columns = ['country', 'description', 'designation', 'province', 'region_1',
       'region_2', 'taster_name', 'taster_twitter_handle', 'title', 'variety',
       'winery']
```

```python
# 문자 영어로 통일 
lists = [["À","Á","Â","Ã","Ä","Ä","Å"]
        ,["à","á","â","ã","ä","ä","å"]
        ,["È","É","Ê","Ë","Ë"]
        ,["è", "é", "ê", "ë", "ë"]     
        ,["ç", "c"]
        ,["ì","í","î","ï","ï"]
        ,["ò","ó","ô","õ","ö","ö","œ"]
        ,["ñ"]
        ,["ù","ú","û","ü","ü"]
        ,["ý","ÿ","ÿ"]]

for list in lists[0][1] :
    df1 = df1.replace(list, "a", regex=True)

for list in lists[2][3] :
    df1 = df1.replace(list, "e", regex=True)
    
for list in lists[4] :
    df1 = df1.replace(list, "c", regex=True)
    
for list in lists[5] :
    df1 = df1.replace(list, "i", regex=True)
    
for list in lists[6] :
    df1 = df1.replace(list, "o", regex=True)
    
for list in lists[7] :
     df1 = df1.replace(list, "n", regex=True)
    
for list in lists[8] :
    df1 = df1.replace(list, "u", regex=True)
    
for list in lists[9] :
    df1 = df1.replace(list, "y", regex=True)
```
