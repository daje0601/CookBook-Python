# Part 1 - Data Preprocessing

## 1.1 분류를 위한 베이스라인 정확도를 계산하세요

## 1.2 모델의 예측 정확도를 시험하기 위한 test set을 만드세요.

> 예측 모델 시험을 위해 **2014-15** ~ **2015-16** 시즌을 `test set`으로 만듭니다. NBA 시즌은 **10월**에 시작해서 **6월**에 종료합니다. 테스트셋에 **1826개**의 데이터가 있다면 알맞게 분할한 것입니다.

```python
df['game_date'] = pd.to_datetime(df['game_date'])
```

```python
condition = df['game_date']>'2014-01-01'
df[condition]
```

```python
train = df[~condition]
test = df[condition]
```

## 1.3 새로운 feature를 만드세요

> 최소 1개 이상의 새로운 feature를 만드세요. 아래의 리스트를 사용하셔도 되고 스스로의 아이디어로 만들어 주셔도 괜찮습니다.
>- **Homecourt Advantage**: 홈 팀(`htm`)이 Los Angeles Lakers (`LAL`)인지 여부
>- **Opponent**: LA Lakers를 상대하는 팀을 나타내는 feature
>- **Seconds remaining in the period**: 피리어드에 남은 총 시간 (분과 초를 합칩니다)
>- **Seconds remaining in the game**: 게임에 남은 총 시간 (피리어드와 초를 합칩니다). 농구에서 게임은 총 4개의 피리어드로 이루어져 있습니다 (각 12분)
>- **Made previous shot**: 코비 브라이언트의 예전 슛이 성공했는지 여부
    

```python
versus = []

def engineer(df):
  # Homecourt Advantage
  df["Homecourt Advantage"] = pd.DataFrame(df["htm"]=="LAL").astype(int)

  # Opponent
  for h_t,v_t in zip(df['htm'],df['vtm']):
    if h_t == "LAL":
      versus.append(v_t)
    else :
      versus.append(h_t)
  df["Opponent"] = pd.DataFrame(versus)

  # Seconds remaining in the period
  df["time_remaining"] = df['minutes_remaining']*60 + df['seconds_remaining']
  df.drop(columns=['minutes_remaining'],inplace = True)
        
  # id 제거, shot_type제거, 시계열 데이터 제거
  df.drop(columns=['game_id','game_event_id','player_name','shot_type','game_date'],inplace = True)

  return df
```

```python
train_eng = engineer(train.copy())
test_eng = engineer(test.copy())
```

```python
features = train_eng.drop(columns=[target]).columns 
X = train_eng[features]
y = train_eng[target]
X_test = test_eng[features]
y_test = test_eng[target]
```

## 1.4 모델을 어떻게 검증할지 결정한 후에 데이터셋을 분할하세요. 

>아래의 검증 방법 중 하나를 선택한 후 목적에 맞게 데이터를 분할 하세요. 어떤 방법을 선택하는지 여부로 채점되지 않습니다 (두 방법 모두 좋은 방법입니다).
>
>- **Train/validate/test split을 통한 검증 (Hold-out Validation): 무작위로 80/20% 비율로 나누기**
>- **독립적인 test set으로 Cross-validation을 통한 검증**. scikit-learn에 있는 모든 cross-validation 방법을 활용하셔도 됩니다.

```python
# Holdout 이용
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(X,y, train_size=0.80,random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from category_encoders import OrdinalEncoder

pipe_tree = make_pipeline(
    OrdinalEncoder(), 
    SimpleImputer(), 
    DecisionTreeClassifier(random_state=2, criterion='entropy',max_depth=5)
)

pipe_tree.fit(X_train, y_train)
print('트리분류 훈련세트 정확도', pipe_tree.score(X_train, y_train))
print('트리분류 훈련세트 f1 score', f1_score(y_train, pipe_tree.predict(X_train)))
```

```python
# 랜덤포레스트
Model = make_pipeline(
    OrdinalEncoder(), 
    SimpleImputer(), 
    RandomForestClassifier(max_depth=6,
                          criterion='entropy',
                          n_jobs=-1,
                          min_samples_leaf=3,
                          random_state=100)
)


Model.fit(X_train, y_train)

print('랜덤 포레스트 훈련 정확도 : ',Model.score(X_train, y_train))
print('랜덤 포레스트 훈련 f1 score : ',f1_score(y_train, Model.predict(X_train)))
```

```python

print('트리분류 훈련세트 정확도', pipe_tree.score(X_train, y_train))
print('트리분류 검증세트 정확도', pipe_tree.score(X_val, y_val))
print('트리분류 훈련세트 f1 score', f1_score(y_train, pipe_tree.predict(X_train)))
print('트리분류 검증세트 f1 score', f1_score(y_val, pipe_tree.predict(X_val)))

print('\n랜덤 포레스트 훈련 정확도 : ',Model.score(X_train, y_train))
print('랜덤 포레스트 검증 정확도 : ',Model.score(X_val, y_val))
print('랜덤 포레스트 훈련 f1 score : ',f1_score(y_train, Model.predict(X_train)))
print('랜덤 포레스트 검증 f1 score : ',f1_score(y_val, Model.predict(X_val)))
```

## Feature Engineering

```python
versus = []

def engineer(df):
  # Homecourt Advantage (1)
  df["Homecourt Advantage"] = pd.DataFrame(df["htm"]=="LAL").astype(int)

  # Opponent (2)
  for h_t,v_t in zip(df['htm'],df['vtm']):
    if h_t == "LAL":
      versus.append(v_t)
    else :
      versus.append(h_t)
  df["Opponent"] = pd.DataFrame(versus)

  # Seconds remaining in the period (3)
  df["period_remaining"] = df['minutes_remaining']*60 + df['seconds_remaining']
  
  # Seconds remaining in the game (4)
  df["game_time_remaining"] = df['period']*60*12 + df['seconds_remaining']
  df.drop(columns=['minutes_remaining'],inplace = True)

  # id 제거, shot_type제거, 시계열 데이터 제거
  df.drop(columns=['game_id','game_event_id','player_name','shot_type','game_date'],inplace = True)

  return df
```

```python
train_eng = engineer(train.copy())
test_eng = engineer(test.copy())

features = train_eng.drop(columns=[target]).columns 
X = train_eng[features]
y = train_eng[target]
X_test = test_eng[features]
y_test = test_eng[target]

X_train,X_val,y_train,y_val = train_test_split(X,y, train_size=0.80,random_state=42)
```

## Optimization

```python
# GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
Model_rcv = make_pipeline(OrdinalEncoder(),
                          SimpleImputer(), 
                          RandomForestClassifier(n_jobs=-1, random_state=100)
  )

dists = {
    "randomforestclassifier__min_samples_leaf": [3,4],
    "randomforestclassifier__max_depth": [6,7,8],  
    "randomforestclassifier__min_samples_split": [2,3],  
}

clf_rcv = RandomizedSearchCV(Model_rcv, 
                      param_distributions=dists, 
                      cv=3
)
clf_rcv.fit(X_train, y_train)
print("최적의 하이퍼 파라미터")
print(clf_rcv.best_params_)
print("f1 score: ", clf_rcv.best_score_)
```

## importance_

```python
# 랜덤포레스트
Model = make_pipeline(
    OrdinalEncoder(), 
    SimpleImputer(), 
    RandomForestClassifier(max_depth=7,
                          criterion='entropy',
                          n_jobs=-1,
                          min_samples_leaf=3,
                          min_samples_split=2,
                          random_state=100)
)


Model.fit(X_train, y_train)

print('랜덤 포레스트 훈련 정확도 : ',Model.score(X_train, y_train))
print('랜덤 포레스트 훈련 f1 score : ',f1_score(y_train, Model.predict(X_train)))
```

```python
# Importance_
zipp  = []
model_rf = Model.named_steps['randomforestclassifier']
for zipper in zip(X_train.columns, model_rf.feature_importances_):
  zipp.append(zipper)
zipp = pd.DataFrame(zipp,columns=['feature','importance']).sort_values('importance',ascending=False)
plt.figure(figsize=(15, 10))
sns.barplot(y = zipp.feature, x= zipp.importance, palette='Blues_r');
```

action type이 가장 높았고 shot_distance가 두번째로 높았습니다 !
