```python
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\user\Documents\pandas_study\data_example\2102_use_time - time.csv", index_col=0)
df
```

```python
columns_sum = df.sum()
df.loc["Jan"] = columns_sum
df.tail()
```

```python
df = df.drop(columns = ["active_time", "sleep_time", "total_time"])
```

```python
df.loc["Jan"].plot(kind="pie", autopct="%0.1f%%", figsize=(6,6), wedgeprops={"alpha":0.4});
```

```python
url = "https://raw.githubusercontent.com/daje0601/pandas_study/main/data_example/2101_use_time%20-%20time.csv"

df = pd.read_csv(url, index_col=0)
df.head()
```

```python
df = df.loc["21.01.25":"21.01.31", : ]
```

```python
columns_sum = df.sum()
df.loc["5week"] = columns_sum
df
```

```python
df = df.drop(columns = ["active_time", "sleep_time", "total_time"])
df.loc["5week"].plot(kind="pie", autopct="%0.1f%%", figsize=(6,6), wedgeprops={"alpha":0.5});
```
